```python
# Import required packages
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dns_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/dns.csv')
dns_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/dns.csv')
ftp_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/ftp.csv')
ftp_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/ftp.csv')
http_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/http.csv')
http_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/http.csv')
lime_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/lime.csv')
lime_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/lime.csv')
localforwarding_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/localforwarding.csv')
localforwarding_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/localforwarding.csv')
remoteforwarding_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/remoteforwarding.csv')
remoteforwarding_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/remoteforwarding.csv')
scp_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/scp.csv')
scp_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/scp.csv')
sftp_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/sftp.csv')
sftp_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/sftp.csv')
shell_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/shell.csv')
shell_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/shell.csv')
telnet_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/telnet.csv')
telnet_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/telnet.csv')
x11_sampled = pd.read_csv('../dataset/dataset-12-features/class-wise-sampled/x11.csv')
x11_original = pd.read_csv('../dataset/dataset-12-features/class-wise-original/x11.csv')
```

```python
def showClusters(name,data1,data2):
    mms = MinMaxScaler()
    data1=data1.drop(['class'],axis=1)
    data2=data2.drop(['class'],axis=1)
    mms.fit(data1)
    data1_transformed = mms.transform(data1)
    mms.fit(data2)
    data2_transformed = mms.transform(data2) 
    Sum_of_squared_distances1 = []
    Sum_of_squared_distances2 = []
    K = range(1,15)
    for k in K:
        km1 = KMeans(n_clusters=k)
        km2 = KMeans(n_clusters=k)
        km1 = km1.fit(data1_transformed)
        km2 = km2.fit(data2_transformed)
        Sum_of_squared_distances1.append(km1.inertia_)
        Sum_of_squared_distances2.append(km2.inertia_)
    plt.plot(K, Sum_of_squared_distances1, 'bx-',label='sampled '+name)
    plt.plot(K, Sum_of_squared_distances2, 'rx-',label='original '+name)
    plt.xlabel('k')
    plt.ylabel('Sum_of_squared_distances')
    plt.legend(loc='upper right')
    plt.title(name)
    plt.show()
```

```python
showClusters('dns',dns_sampled,dns_original)
```

```python
showClusters('ftp',ftp_sampled,ftp_original)
```

```python
showClusters('http',http_sampled,http_original)
```
